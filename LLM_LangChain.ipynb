### O que é LangChain (visão moderna)

LangChain é um framework para conectar LLMs com dados, ferramentas e lógica de aplicação.

Ele permite:

* conectar modelos (OpenAI, Claude, Llama)

* usar dados externos

* criar pipelines de IA

* criar agentes que usam ferramentas

Casos comuns:

* Chatbots

* RAG (chat com documentos)

* automação com agentes

* análise de dados

* copilots internos

| Componente       | Função                        |
| ---------------- | ----------------------------- |
| Models           | LLM ou embeddings             |
| Prompt Templates | templates de prompt           |
| LCEL             | nova linguagem para pipelines |
| Tools            | funções externas              |
| Agents           | IA que decide ações           |
| LangGraph        | agentes complexos             |

```python
# Agora vamos separar em função para poder usar dentro do LangChain:

from dotenv import load_dotenv
load_dotenv()

import os
from huggingface_hub import InferenceClient

client = InferenceClient(
    api_key=os.getenv("HUGGINGFACEHUB_API_TOKEN")
)

def call_hf_chat(user_text: str)->str:
    response = client.chat_completion(
        model="meta-llama/Meta-Llama-3-8B-Instruct",
        messages=[
            {"role": "user",
             "content": user_text}
        ],
        max_tokens=2_000,
        temperature=0.7
    )
    return response.choices[0].message.content

print(call_hf_chat("Explique a equação de schrodinger, de maneira simples!"))
```

```python
# conectar  modelo ao Prompt Template usando Runnable


from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableLambda

prompt = ChatPromptTemplate.from_messages([
    ("system", "Você é um professor claro e objetivo."),
    ("user", "Explique {tema} em linguagem simples, com no máximo 3 frases.")
])

def hf_runnable(prompt_value):
    messages = []
    for msg in prompt_value.to_messages():
        messages.append({
            "role": msg.type if msg.type != "human" else "user",
            "content": msg.content
        })

    response = client.chat_completion(
        model="meta-llama/Meta-Llama-3-8B-Instruct",
        messages=messages,
        max_tokens=128,
        temperature=0.7,
    )
    return response.choices[0].message.content

chain = prompt | RunnableLambda(hf_runnable)

resultado = chain.invoke({"tema": "Física Moderna"})
print(resultado)
```

```python
# CHAIN: uma sequência de etapas que processam uma entrada até gerar uma saída.

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableLambda

prompt = ChatPromptTemplate.from_messages([
    ("system", "Você é um especialista em IA e responde em português do Brasil."),
    ("user", "Explique {tema} para um estudante iniciante e depois dê 2 exemplos práticos.")
])

def hf_chat(prompt_value):
    messages = []
    for msg in prompt_value.to_messages():
        role = "user" if msg.type == "human" else msg.type
        messages.append({"role": role, "content": msg.content})

    response = client.chat_completion(
        model="meta-llama/Meta-Llama-3-8B-Instruct",
        messages=messages,
        max_tokens=2_000,
        temperature=0.7,
    )
    return response.choices[0].message.content

chain = prompt | RunnableLambda(hf_chat)

print(chain.invoke({"tema": "Prompt Engineering"}))
```

```python
"""
Encadeamento de etapas

A documentação atual também mostra workflows em que uma etapa usa a saída da anterior. 
Para tarefas bem definidas, isso é tratado como prompt chaining.
"""

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableLambda

def hf_chat(prompt_value):
    messages = []
    for msg in prompt_value.to_messages():
        role = "user" if msg.type == "human" else msg.type
        messages.append({"role": role, "content": msg.content})

    response = client.chat_completion(
        model="meta-llama/Meta-Llama-3-8B-Instruct",
        messages=messages,
        max_tokens=2_000,
        temperature=0.7,
    )
    return response.choices[0].message.content

resumir_prompt = ChatPromptTemplate.from_messages([
    ("system", "Você resume textos com clareza."),
    ("user", "Resuma o tema {tema} em 4 linhas.")
])

topicos_prompt = ChatPromptTemplate.from_messages([
    ("system", "Você organiza conteúdo em tópicos."),
    ("user", "Transforme este resumo em 5 tópicos objetivos:\n\n{resumo}")
])

resumir_chain = resumir_prompt | RunnableLambda(hf_chat)

topicos_chain = (
    {"resumo": resumir_chain}
    | topicos_prompt
    | RunnableLambda(hf_chat)
)

print(topicos_chain.invoke({"tema": "prompt chaining"}))
```

```python
from dotenv import load_dotenv
load_dotenv()

import os
from huggingface_hub import InferenceClient
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableLambda

client = InferenceClient(
    api_key=os.getenv("HUGGINGFACEHUB_API_TOKEN")
)

prompt = ChatPromptTemplate.from_messages([
    ("system", "Você é um tutor de programação. Responda em português."),
    ("user", "Explique {tema} com definição, exemplo e dica prática.")
])

def hf_chat(prompt_value):
    messages = []
    for msg in prompt_value.to_messages():
        role = "user" if msg.type == "human" else msg.type
        messages.append({"role": role, "content": msg.content})

    response = client.chat_completion(
        model="meta-llama/Meta-Llama-3-8B-Instruct",
        messages=messages,
        max_tokens=2_000,
        temperature=0.7,
    )
    return response.choices[0].message.content

chain = prompt | RunnableLambda(hf_chat)

resultado = chain.invoke({"tema": "LangChain"})
print(resultado)
```
