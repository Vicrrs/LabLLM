# RAG - Retrieval Augmented Generation

RAG é uma técnica que combina modelos de linguagem com mecanismos de recuperação de informações para melhorar a geração de texto

Em vez de confiar apenas nos dados armazenados internamente pelo modelo, RAG recupera informações relevantes de fontes externas, como documentos ou bancos de dados para fornecer

```python
# Teste de indexação no FAISS
from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

def test_faiss_index():
    docs = [
        Document(page_content="LangChain é um framework para LLMs."),
        Document(page_content="RAG significa Retrieval Augmented Generation."),
        Document(page_content="LangChain conecta LLMs a fontes externas.")
    ]

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstore = FAISS.from_documents(docs, embeddings)

    assert vectorstore is not None
    print("Índice FAISS criado com sucesso")

test_faiss_index()
```

```python
# Teste de recuperação
from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

def test_retriever():
    docs = [
        Document(page_content="LangChain é um framework para LLMs."),
        Document(page_content="RAG significa Retrieval Augmented Generation."),
        Document(page_content="LangChain conecta LLMs a fontes externas.")
    ]

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstore = FAISS.from_documents(docs, embeddings)
    retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

    results = retriever.invoke("O que é RAG?")

    assert len(results) > 0
    print("✅ Recuperação funcionando")
    print("\nDocumentos recuperados:")
    for i, doc in enumerate(results, 1):
        print(f"{i}. {doc.page_content}")

test_retriever()
```

```python
# Teste de montagem do contexto

from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

def test_context_building():
    docs = [
        Document(page_content="LangChain é um framework para LLMs."),
        Document(page_content="RAG significa Retrieval Augmented Generation."),
        Document(page_content="LangChain conecta LLMs a fontes externas.")
    ]

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstore = FAISS.from_documents(docs, embeddings)
    retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

    results = retriever.invoke("O que é RAG?")
    context = "\n".join([d.page_content for d in results])

    assert isinstance(context, str)
    assert len(context) > 0
    print("✅ Contexto montado com sucesso")
    print("\nContexto:\n")
    print(context)

test_context_building()
```

```python
# Teste do prompt
from langchain_core.prompts import ChatPromptTemplate

def test_prompt():
    prompt = ChatPromptTemplate.from_template(
        """
Use o contexto abaixo para responder a pergunta.

Contexto:
{context}

Pergunta:
{question}
"""
    )

    final_prompt = prompt.format(
        context="RAG significa Retrieval Augmented Generation.",
        question="O que é RAG?"
    )

    assert "RAG significa" in final_prompt
    assert "O que é RAG?" in final_prompt
    print("✅ Prompt formatado com sucesso")
    print("\nPrompt final:\n")
    print(final_prompt)

test_prompt()
```

```python
# Teste do pipeline RAG com mock

from langchain_core.documents import Document
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

def fake_llm(messages):
    return "Resposta simulada: RAG é uma técnica que combina recuperação de documentos com geração de texto."

def test_rag_with_mock():
    docs = [
        Document(page_content="LangChain é um framework para LLMs."),
        Document(page_content="RAG significa Retrieval Augmented Generation."),
        Document(page_content="LangChain conecta LLMs a fontes externas.")
    ]

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstore = FAISS.from_documents(docs, embeddings)
    retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

    prompt = ChatPromptTemplate.from_template(
        """
Use o contexto abaixo para responder a pergunta.

Contexto:
{context}

Pergunta:
{question}
"""
    )

    def rag(question):
        results = retriever.invoke(question)
        context = "\n".join([d.page_content for d in results])

        messages = [
            {
                "role": "user",
                "content": prompt.format(context=context, question=question)
            }
        ]
        return fake_llm(messages)

    answer = rag("O que é RAG?")
    assert "RAG" in answer
    print(" Pipeline RAG com mock funcionando")
    print("\nResposta:\n")
    print(answer)

test_rag_with_mock()
```

```python
# Teste real com Hugging Face

from dotenv import load_dotenv
load_dotenv()

import os
from huggingface_hub import InferenceClient
from langchain_core.documents import Document
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

def test_rag_with_hf():
    client = InferenceClient(
        api_key=os.getenv("HUGGINGFACEHUB_API_TOKEN")
    )

    docs = [
        Document(page_content="LangChain é um framework para construir aplicações com LLMs."),
        Document(page_content="RAG significa Retrieval Augmented Generation."),
        Document(page_content="LangChain permite conectar LLMs com dados externos.")
    ]

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstore = FAISS.from_documents(docs, embeddings)
    retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

    prompt = ChatPromptTemplate.from_template(
        """
Responda com base apenas no contexto abaixo.

Contexto:
{context}

Pergunta:
{question}
"""
    )

    def hf_chat(messages):
        response = client.chat_completion(
            model="meta-llama/Meta-Llama-3-8B-Instruct",
            messages=messages,
            max_tokens=200,
            temperature=0.2,
        )
        return response.choices[0].message.content

    def rag(question):
        results = retriever.invoke(question)
        context = "\n".join([d.page_content for d in results])

        messages = [
            {
                "role": "user",
                "content": prompt.format(context=context, question=question)
            }
        ]
        return hf_chat(messages)

    answer = rag("O que é RAG?")
    print("Pipeline RAG com Hugging Face funcionando")
    print("\nResposta:\n")
    print(answer)

test_rag_with_hf()
```

```python
# Teste de depuração

from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

def debug_rag(question):
    docs = [
        Document(page_content="LangChain é um framework para construir aplicações com LLMs."),
        Document(page_content="RAG significa Retrieval Augmented Generation."),
        Document(page_content="LangChain permite conectar LLMs com dados externos.")
    ]

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstore = FAISS.from_documents(docs, embeddings)
    retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

    results = retriever.invoke(question)

    print("Pergunta:", question)
    print("\nDocumentos recuperados:")
    for i, doc in enumerate(results, 1):
        print(f"{i}. {doc.page_content}")

    context = "\n".join([d.page_content for d in results])

    print("\nContexto final:")
    print(context)

debug_rag("O que é RAG?")
```
